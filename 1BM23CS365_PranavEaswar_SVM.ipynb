Load the 'letter-recognition.xlsx' file into a pandas DataFrame and inspect its structure. Note: The dataset has 16 features (A-P) and the first column 'letter' is the target variable.

## Load Letter-recognition Dataset

### Subtask:
Load the 'letter-recognition.xlsx' file into a pandas DataFrame and inspect its structure. Note: The dataset has 16 features (A-P) and the first column 'letter' is the target variable.


**Reasoning**:
To load the 'letter-recognition.xlsx' file, I will use `pd.read_excel()`. Then, I will display the first 5 rows with `.head()` and print a concise summary using `.info()` to inspect its structure as requested.



```python
import pandas as pd

letter_recognition_df = pd.read_excel('/content/letter-recognition.xlsx')

print("First 5 rows of the Letter-recognition DataFrame:")
print(letter_recognition_df.head())

print("\nDataFrame Info:")
letter_recognition_df.info()
```

## Prepare Letter-recognition Data

### Subtask:
Separate features (X) and target (y) from the Letter-recognition dataset. Encode the target variable if necessary. Then, split the data into 80% training and 20% testing sets.


**Reasoning**:
The subtask requires separating features and target, encoding the target variable, and splitting the data into training and testing sets. I will perform these steps in a single code block for efficiency.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Separate features (X) and target (y)
X_letter = letter_recognition_df.drop('letter', axis='columns')
y_letter = letter_recognition_df['letter']

# Encode the target variable
label_encoder = LabelEncoder()
y_letter_encoded = label_encoder.fit_transform(y_letter)

# Split the data into training and testing sets
X_train_letter, X_test_letter, y_train_letter, y_test_letter = train_test_split(X_letter, y_letter_encoded, test_size=0.2, random_state=42)

print(f"X_train_letter shape: {X_train_letter.shape}")
print(f"X_test_letter shape: {X_test_letter.shape}")
print(f"y_train_letter shape: {y_train_letter.shape}")
print(f"y_test_letter shape: {y_test_letter.shape}")
```

## Train and Evaluate RBF SVM for Letter-recognition

### Subtask:
Build an SVM classifier using the RBF kernel. Train the model on the training data, then predict on the test data. Display the accuracy score and the confusion matrix. Plot the ROC curve and display the AUC score.


**Reasoning**:
The subtask requires building an SVM classifier with an RBF kernel, training it, making predictions, and then calculating and displaying the accuracy score, confusion matrix, and ROC AUC score. This code block imports the necessary modules, initializes the SVC model with an RBF kernel and `probability=True`, trains it on the Letter-recognition training data, makes predictions and gets probabilities on the test data, and then computes and prints the accuracy score, confusion matrix, and weighted average ROC AUC score.



```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.preprocessing import LabelBinarizer
import matplotlib.pyplot as plt

# Initialize SVC model with RBF kernel and probability=True
rbf_svc_letter = SVC(kernel='rbf', random_state=42, probability=True)

# Train the model
rbf_svc_letter.fit(X_train_letter, y_train_letter)

# Make predictions on the test data
y_pred_rbf_letter = rbf_svc_letter.predict(X_test_letter)

# Calculate and print accuracy score
print(f"Accuracy Score (RBF Kernel): {accuracy_score(y_test_letter, y_pred_rbf_letter):.4f}")

# Calculate and print confusion matrix
print("\nConfusion Matrix (RBF Kernel):\n", confusion_matrix(y_test_letter, y_pred_rbf_letter))

# Calculate multiclass ROC AUC score
# Instantiate LabelBinarizer and fit-transform y_test_letter
label_binarizer = LabelBinarizer()
y_test_letter_binarized = label_binarizer.fit_transform(y_test_letter)

# Get predicted probabilities for each class
y_pred_proba_rbf_letter = rbf_svc_letter.predict_proba(X_test_letter)

# Calculate weighted average ROC AUC score
roc_auc = roc_auc_score(y_test_letter_binarized, y_pred_proba_rbf_letter, multi_class='ovr', average='weighted')
print(f"\nWeighted Average ROC AUC Score (RBF Kernel): {roc_auc:.4f}")
```

## Train and Evaluate Linear SVM for Letter-recognition

### Subtask:
Build an SVM classifier using the linear kernel. Train the model on the training data, then predict on the test data. Display the accuracy score and the confusion matrix. Plot the ROC curve and display the AUC score.


**Reasoning**:
The subtask requires building an SVM classifier with a linear kernel, training it, making predictions, and then calculating and displaying the accuracy score, confusion matrix, and ROC AUC score. This code block initializes the SVC model with a linear kernel and `probability=True`, trains it on the Letter-recognition training data, makes predictions and gets probabilities on the test data, and then computes and prints the accuracy score, confusion matrix, and weighted average ROC AUC score.



```python
linear_svc_letter = SVC(kernel='linear', random_state=42, probability=True)

linear_svc_letter.fit(X_train_letter, y_train_letter)

y_pred_linear_letter = linear_svc_letter.predict(X_test_letter)

print(f"Accuracy Score (Linear Kernel): {accuracy_score(y_test_letter, y_pred_linear_letter):.4f}")
print("\nConfusion Matrix (Linear Kernel):\n", confusion_matrix(y_test_letter, y_pred_linear_letter))

y_pred_proba_linear_letter = linear_svc_letter.predict_proba(X_test_letter)

roc_auc_linear = roc_auc_score(y_test_letter_binarized, y_pred_proba_linear_letter, multi_class='ovr', average='weighted')
print(f"\nWeighted Average ROC AUC Score (Linear Kernel): {roc_auc_linear:.4f}")
```

**Reasoning**:
To complete the subtask of plotting the ROC curve, I will generate a code block that iterates through each class, computes the ROC curve and AUC for each, and then plots them on a single graph.



```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get the number of classes
n_classes = len(label_binarizer.classes_)

# Plot ROC curve for each class
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_letter_binarized[:, i], y_pred_proba_linear_letter[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve of class {label_binarizer.classes_[i]} (area = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Linear Kernel')
plt.legend(loc='lower right')
plt.show()
```
